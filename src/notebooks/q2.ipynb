## Question 2 - PCA

PCA can be calculated by following these steps
1. Standardize the data
   
   $ x_{ij}' = \frac{x_{ij} - \mu_j}{\sigma_j} $
   
2. Compute covariance matrix
   
   $  C = \frac{1}{n - 1} X'^{\top} X' $

3. Compute eigenvalues and eigenvectors
   
   $ C v_k = \lambda_k v_k $

4. Sort and select principal compenents
   
   $ V_k = \text{Sorted eigenvectors corresponding to top } k \text{ eigenvalues} $

5. Transform the data
   
   $ Z = X' V_k $
